### grid function
1. https://www.w3schools.com/python/matplotlib_grid.asp

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Plot of Points')
# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Plot of Points')

# plt.grid() # or Mention true 
# plt.grid(b=True)
# plt.grid(color = 'red')
plt.grid(axis='x')  # change with y 
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Plot of Points')

plt.minorticks_on()
plt.grid()  
plt.show()
```
