```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('number')
plt.ylabel('numbers square')
plt.title('square  numbers')
```

```python
#  if we give only one axis value then it is by default considered as a y axis 
plt.plot([1,4,9,16,25])
plt.xlabel("n")
plt.ylabel("n*n")
```

```python
plt.plot([1,4,9,16,25], "g")    # g means line should be of green color
plt.xlabel("n")
plt.ylabel("n*n")
```

```python
plt.plot([1,4,9,16,25], "r")    # r means line should be of red color
plt.xlabel("n")
plt.ylabel("n*n")
```

```python
plt.plot([1,4,9,16,25], "g+")    # g+ means line should be of green color with + as marker
plt.xlabel("n")
plt.ylabel("n*n")
```

```python
plt.plot([1,4,9,16,25], "g+-")    # g+ means line should be of green color with + as marker
plt.xlabel("n")
plt.ylabel("n*n")
```

```python
plt.plot([1,4,9,16,25], "g+", markersize = 20)    # markersize 
plt.xlabel("n")
plt.ylabel("n*n")
```

```python
plt.plot([1,4,9,16,25], "g", linewidth = 10)    # linewidth
plt.xlabel("n")
plt.ylabel("n*n")
```

```python
plt.plot([1,4,9,16,25], color = 'red')    # linewidth
plt.xlabel("n")
plt.ylabel("n*n")
```

```python
# when there is a confilict between format string and keyword argument, keyword argument has higher priority
plt.plot([1,4,9,16,25], 'g', color = 'red')    # conflict b/w red and green color
```

```python
plt.plot([10,20,30,40], "r")
plt.plot([50,100,150,200], "g")
plt.show()
```

```python
plt.plot([10,20,30,40], "r", [50,100,150,200], "g")
plt.show()
```

    - learnings 
    1. format string is used to decorate the groph 
    2. it is optional if not provided then default values are taken
    3. format string consist of three parts which are (marker, line, color) or (color, marker, line)
    4. default color is blue and marker is solid i.e 'b-', for sequence of data it is b, o, g
