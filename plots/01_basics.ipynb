    - keyword arguments 
    1. used to change the look of graph 
    2. line width - used to change the width of the line
    3. antialiased - smoothen the line of the plot, it is a boolean value 
    4. color or c - used to specify the color of plot markers
    5. dashes - to change the style of dash style line 
    6. linestyle or ls -> To change the style of line 
    7. marker - to display the break point
    8. marker edge color - to change the color of marker edge
    9. marker edge width - to change the thickness of the marker
    10. markersize - to change the size of marker
    11. markerfacecolor - to change the color of marker face
    12. markevery - Tells about where to use marker

```python
# Assignment - 01
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(14)
y = np.sin(x / 2)

plt.step(x, y + 2, label='pre (default)')
plt.plot(x, y + 2, 'o--', color='grey', alpha=0.3)

plt.step(x, y + 1, where='mid', label='mid')
plt.plot(x, y + 1, 'o--', color='grey', alpha=0.3)

plt.step(x, y, where='post', label='post')
plt.plot(x, y, 'o--', color='grey', alpha=0.3)

plt.grid(axis='x', color='0.95')
plt.legend(title='Parameter where:')
plt.title('plt.step(where=...)')
plt.show()
```

    - keyword arguments (alpha and zorder)
    1. used to change the look of graph 
    2. line width - used to change the width of the line
    3. antialiased - smoothen the line of the plot, it is a boolean value 
    4. color or c - used to specify the color of plot markers
    5. dashes - to change the style of dash style line 
    6. linestyle or ls -> To change the style of line 
    7. marker - to display the break point
    8. marker edge color - to change the color of marker edge
    9. marker edge width - to change the thickness of the marker
    10. markersize - to change the size of marker
    11. markerfacecolor - to change the color of marker face
    12. markevery - Tells about where to use marker
    13. alpha - using this we can adjust the transperancy of the graph, values can be between 0 to 1, default value is 1
    14. zorder - used to decide which plot is to show upfront and which plot is need to send back (this is useful when we are ploting multiple set of data in single plot)

```python
a = np.arange(0, 10, 0.5)
plt.plot(a, alpha = 0.5)
plt.plot(a**2, alpha = 0.2)
plt.plot(a**3, alpha = 0.7)
plt.show()
```

```python
a = np.arange(0, 10, 0.5)
plt.plot(a, alpha = 0.5, linewidth = 5)
plt.plot(a**2, alpha = 0.2,linewidth = 5)
plt.plot(a**3, alpha = 0.7, linewidth = 5)
plt.show()
```

```python
plt.plot([10,20,80,90], linewidth = 10 )
plt.plot([10,20,30,40],linewidth = 10)
plt.show()
```

```python
plt.plot([10,20,80,90],linewidth = 10, zorder = 100,)
plt.plot([10,20,30,40],linewidth = 10, zorder = 50)
plt.show()
# the plot with higher value of zorder or z-order will be in upfront
```

```python
# to get more pretty curve use it with alpha 
plt.plot([10,20,80,90],linewidth = 10, zorder = 100, alpha = 0.5)
plt.plot([10,20,30,40],linewidth = 10, zorder = 50)
plt.show()
```

```python
plt.plot([0,0], [-1,1], linewidth=10, zorder=100, alpha = 0.7)
plt.plot([-1,1], [-1,1], linewidth=10, zorder = 90, alpha = 0.5)
plt.plot([-1,1], [1,-1], linewidth=10, zorder = 80 )
plt.show()
```

```python
a = np.arange(0, 10, 0.5)
plt.plot(a**2)
plt.xlabel('number', {'size': 20, 'color': 'blue'})
plt.ylabel('squared number')
plt.show()
```

```python
a = np.arange(0, 10, 0.5)
plt.plot(a**2)
plt.xlabel('number', {'size': 20, 'color': 'blue'}, labelpad = 20)
# labelpad is the distance between the axis and the label
# positive values moves toward the bottom 
# negative values moves toward the top
plt.ylabel('squared number')
plt.show()
```

```python
# fontsize values available 
a = np.arange(0, 10, 0.5)
plt.plot(a**2)
plt.xlabel('number', {'size': 20})
plt.ylabel('squared number', size = 'xx-small')   # size argument can take numeric values also like 20,30,10
plt.show()
# note the two ways to define the size of x and y label
```

```python
# alpha keyword is also availble for x and y label
a = np.arange(0, 10, 0.5)
plt.plot(a**2)
plt.xlabel('number', {'size': 20, 'color': 'red'}, alpha=0.5)
plt.ylabel('squared number', size = 20)  
plt.show()
```

```python
# background color 
a = np.arange(0, 10, 0.5)
plt.plot(a**2)
#plt.xlabel('number', {'size': 20, 'color': 'red'}, backgroundcolor = 'blue')
plt.xlabel('number', {'size': 20, 'color': 'red', 'backgroundcolor': 'blue'})
plt.ylabel('squared number', size = 20, backgroundcolor = 'orange')  
plt.show()
```

```python
# background color 
a = np.arange(0, 10, 0.5)
plt.plot(a**2)
#plt.xlabel('number', {'size': 20, 'color': 'red'}, backgroundcolor = 'blue')
plt.xlabel('number', {'size': 20, 'color': 'red', 'backgroundcolor': 'blue'})
plt.ylabel('squared number', size = 20, backgroundcolor = 'orange')  
plt.title('square of the number')
plt.show()
```
